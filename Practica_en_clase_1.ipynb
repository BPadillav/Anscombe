En esta celda se importan las libreras necesarias para el anlisis de datos y la visualizacin, incluyendo pandas para el manejo de datos, numpy para operaciones numricas, matplotlib y seaborn para generar grficas.

```python
#Importacin de mdulos necesarios
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt 
import matplotlib as mpl
import seaborn as sns
import copy

print("Mdulos y clases importados")
```

En esta celda se carga el dataset df_anscombe.csv y se genera un resumen estadstico de cada grupo, utilizando la funcin groupby para agrupar los datos por la columna group.

```python
#Carga del dataset y generacin de resmenes estadsticos por grupo
ruta = 'df_anscombe.csv'  
df_anscombe = pd.read_csv(ruta)

# Resumen estadstico agrupado por la columna 'group'
df_anscombe.groupby('group').describe()
```

En esta celda se generan grficas de dispersin para cada grupo de datos. Se itera sobre cada grupo nico en la columna group y se crea una grfica de dispersin con los valores de x e y.

```python
#Generacin de grficas de dispersin (scatterplot) por cada grupo

# Cargamos el dataset
ruta_dataset = 'df_anscombe.csv' 
df_anscombe = pd.read_csv(ruta_dataset)

# Extraemos los grupos nicos en el dataset
grupos = df_anscombe['group'].unique()

# Iteramos sobre cada grupo y creamos una grfica de dispersin para cada uno
for group in grupos:
    subset = df_anscombe[df_anscombe['group'] == group]  # Filtramos por grupo
    
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x='x', y='y', data=subset, s=100, color='purple', edgecolor='black')  # Grfico de dispersin
    
    plt.title(f'Grupo {group}')  # Ttulo de la grfica
    plt.xlabel('X')  # Etiqueta eje X
    plt.ylabel('Y')  # Etiqueta eje Y
    plt.grid(True)  # Activar cuadrcula
    plt.show()  # Mostrar grfica
```

En esta celda se crean dos grficos de box-plot, uno para la variable x y otro para la variable y, distribuidos por los grupos del dataset. El grfico de box-plot permite visualizar la distribucin y detectar valores atpicos.



```python
#Creacin de box-plots para la distribucin de las variables X y Y por grupo

# Cargar dataset
ruta = 'df_anscombe.csv' 
df_anscombe = pd.read_csv(ruta)

# Estilo de Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(14, 6))

# Subplot para el boxplot de X por grupo
plt.subplot(1, 2, 1) 
sns.boxplot(x='group', y='x', data=df_anscombe)
plt.title('Distribucin de X por Grupo')
plt.xlabel('Grupo')
plt.ylabel('X')

# Subplot para el boxplot de Y por grupo
plt.subplot(1, 2, 2)  
sns.boxplot(x='group', y='y', data=df_anscombe)
plt.title('Distribucin de Y por Grupo')
plt.xlabel('Grupo')
plt.ylabel('Y')

# Ajustar y mostrar las grficas
plt.tight_layout()
plt.show()
```
