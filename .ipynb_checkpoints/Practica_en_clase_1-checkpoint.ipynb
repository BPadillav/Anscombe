En esta celda se importan las librerías necesarias para el análisis de datos y la visualización, incluyendo pandas para el manejo de datos, numpy para operaciones numéricas, matplotlib y seaborn para generar gráficas.

```python
#Importación de módulos necesarios
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt 
import matplotlib as mpl
import seaborn as sns
import copy

print("Módulos y clases importados")
```

En esta celda se carga el dataset df_anscombe.csv y se genera un resumen estadístico de cada grupo, utilizando la función groupby para agrupar los datos por la columna group.

```python
#Carga del dataset y generación de resúmenes estadísticos por grupo
ruta = 'df_anscombe.csv'  
df_anscombe = pd.read_csv(ruta)

# Resumen estadístico agrupado por la columna 'group'
df_anscombe.groupby('group').describe()
```

En esta celda se generan gráficas de dispersión para cada grupo de datos. Se itera sobre cada grupo único en la columna group y se crea una gráfica de dispersión con los valores de x e y.

```python
#Generación de gráficas de dispersión (scatterplot) por cada grupo

# Cargamos el dataset
ruta_dataset = 'df_anscombe.csv' 
df_anscombe = pd.read_csv(ruta_dataset)

# Extraemos los grupos únicos en el dataset
grupos = df_anscombe['group'].unique()

# Iteramos sobre cada grupo y creamos una gráfica de dispersión para cada uno
for group in grupos:
    subset = df_anscombe[df_anscombe['group'] == group]  # Filtramos por grupo
    
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x='x', y='y', data=subset, s=100, color='purple', edgecolor='black')  # Gráfico de dispersión
    
    plt.title(f'Grupo {group}')  # Título de la gráfica
    plt.xlabel('X')  # Etiqueta eje X
    plt.ylabel('Y')  # Etiqueta eje Y
    plt.grid(True)  # Activar cuadrícula
    plt.show()  # Mostrar gráfica
```

En esta celda se crean dos gráficos de box-plot, uno para la variable x y otro para la variable y, distribuidos por los grupos del dataset. El gráfico de box-plot permite visualizar la distribución y detectar valores atípicos.



```python
#Creación de box-plots para la distribución de las variables X y Y por grupo

# Cargar dataset
ruta = 'df_anscombe.csv' 
df_anscombe = pd.read_csv(ruta)

# Estilo de Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(14, 6))

# Subplot para el boxplot de X por grupo
plt.subplot(1, 2, 1) 
sns.boxplot(x='group', y='x', data=df_anscombe)
plt.title('Distribución de X por Grupo')
plt.xlabel('Grupo')
plt.ylabel('X')

# Subplot para el boxplot de Y por grupo
plt.subplot(1, 2, 2)  
sns.boxplot(x='group', y='y', data=df_anscombe)
plt.title('Distribución de Y por Grupo')
plt.xlabel('Grupo')
plt.ylabel('Y')

# Ajustar y mostrar las gráficas
plt.tight_layout()
plt.show()
```
